```python
## Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import dataframe_image as dfi
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
## Loading in the QoG dataframe
qog = pd.read_csv('data/qog_std_ts_jan20.csv')
qog.head()
```

Fixing the variables within the DF (I got the information about their metadata by loading them into STATA). Included the folder as STATAFiles.

```python
## Appointment for Life to Constitutional Court
qog.iaep_alcc  = qog.iaep_alcc.replace('.a', np.NaN) ## Missing Information
qog.iaep_alcc  = qog.iaep_alcc.replace('.b', np.NaN) ## no constitution
qog.iaep_alcc  = qog.iaep_alcc.replace('.g', np.NaN) ## no const. court

##  Appointment of Regional Representatives
qog.iaep_arr  = qog.iaep_arr.replace('.a', np.NaN) ## Missing Information

## Constitutional Court
qog.iaep_cc  = qog.iaep_cc.replace('.a', np.NaN) ## Missing Information
qog.iaep_cc  = qog.iaep_cc.replace('.b', np.NaN) ## no constitution

##  Constitutional Court Rules on Executive Actions
qog.iaep_ccrea  = qog.iaep_ccrea.replace('.a', np.NaN) ## Missing Information
qog.iaep_ccrea  = qog.iaep_ccrea.replace('.b', np.NaN) ## no constitution
qog.iaep_ccrea  = qog.iaep_ccrea.replace('.g', np.NaN) ## no const. court

## Constitutional Court Rules on Legislative Actions
qog.iaep_ccrla  = qog.iaep_ccrla.replace('.a', np.NaN) ## Missing Information
qog.iaep_ccrla  = qog.iaep_ccrla.replace('.b', np.NaN) ## no constitution
qog.iaep_ccrla  = qog.iaep_ccrla.replace('.g', np.NaN) ## no const. court
qog.iaep_ccrla  = qog.iaep_ccrla.replace('.e', np.NaN) ## no legislature

##  Executive Can Change Domestic Taxes
qog.iaep_eccdt  = qog.iaep_eccdt.replace('.a', np.NaN) ## Missing Information
qog.iaep_eccdt  = qog.iaep_eccdt.replace('.e', np.NaN) ## no legislature

## Executive Can Dissolve Legislature
qog.iaep_ecdl  = qog.iaep_ecdl.replace('.a', np.NaN) ## Missing Information
qog.iaep_ecdl  = qog.iaep_ecdl.replace('.b', np.NaN) ## no constitution
qog.iaep_ecdl  = qog.iaep_ecdl.replace('.e', np.NaN) ## no legislature

## Election of the Executive
qog.iaep_ee  = qog.iaep_ee.replace('.a', np.NaN) ## Missing Information

## Executive is Member of Legislature
qog.iaep_eml  = qog.iaep_eml.replace('.a', np.NaN) ## Missing Information
qog.iaep_eml  = qog.iaep_eml.replace('.e', np.NaN) ## no legislature

## Executive Power over Military Force
qog.iaep_epmf  = qog.iaep_epmf.replace('.a', np.NaN) ## Missing Information
qog.iaep_epmf  = qog.iaep_epmf.replace('.e', np.NaN) ## no legislature

## Electoral System
qog.iaep_es  = qog.iaep_es.replace('.a', np.NaN) ## Missing Information
qog.iaep_es  = qog.iaep_es.replace('.e', np.NaN) ## no legislature
qog.iaep_es  = qog.iaep_es.replace('.l', np.NaN) ## no elected legislature

## Electoral System for the Executive
qog.iaep_ese  = qog.iaep_ese.replace('.a', np.NaN) ## Missing Information
qog.iaep_ese  = qog.iaep_ese.replace('.j', np.NaN) ## no elected executive

## Executive Veto Power
qog.iaep_evp  = qog.iaep_evp.replace('.a', np.NaN) ## Missing Information
qog.iaep_evp  = qog.iaep_evp.replace('.e', np.NaN) ## no legislature
qog.iaep_evp  = qog.iaep_evp.replace('.b', np.NaN) ## no constitution

## Independence of Selection of Executive
qog.iaep_ise  = qog.iaep_ise.replace('.a', np.NaN) ## Missing Information
qog.iaep_ise  = qog.iaep_ise.replace('.e', np.NaN) ## no legislature

## Legislature Approves Budget
qog.iaep_lap  = qog.iaep_lap.replace('.a', np.NaN) ## Missing Information
qog.iaep_lap  = qog.iaep_lap.replace('.e', np.NaN) ## no legislature

## Legislature Can Remove Executive
qog.iaep_lcre  = qog.iaep_lcre.replace('.a', np.NaN) ## Missing Information
qog.iaep_lcre  = qog.iaep_lcre.replace('.e', np.NaN) ## no legislature
qog.iaep_lcre  = qog.iaep_lcre.replace('.b', np.NaN) ## no constitution

## Legislature Veto Power
qog.iaep_lvp  = qog.iaep_lvp.replace('.a', np.NaN) ## Missing Information
qog.iaep_lvp  = qog.iaep_lvp.replace('.e', np.NaN) ## no legislature
qog.iaep_lvp  = qog.iaep_lvp.replace('.b', np.NaN) ## no constitution

## National Elections for an Executive
qog.iaep_nee  = qog.iaep_nee.replace('.a', np.NaN) ## Missing Information

## National Elections for the Legislature
qog.iaep_nel  = qog.iaep_nel.replace('.a', np.NaN) ## Missing Information
qog.iaep_nel  = qog.iaep_nel.replace('.e', np.NaN) ## no legislature

## National Referendums
qog.iaep_nr  = qog.iaep_nr.replace('.a', np.NaN) ## Missing Information

## Parties with More than 5 Percent
qog.iaep_pm5p  = qog.iaep_pm5p.replace('.a', np.NaN) ## Missing Information
qog.iaep_pm5p  = qog.iaep_pm5p.replace('.e', np.NaN) ## no legislature

## Removal of Members of Constitutional Court
qog.iaep_rmcc  = qog.iaep_rmcc.replace('.a', np.NaN) ## Missing Information
qog.iaep_rmcc  = qog.iaep_rmcc.replace('.b', np.NaN) ## no constitution
qog.iaep_rmcc  = qog.iaep_rmcc.replace('.g', np.NaN) ## no const. court

## Unitary or Federal State
qog.iaep_ufs  = qog.iaep_ufs.replace('.a', np.NaN) ## Missing Information

## The Age of the Constitution (years)
qog.iaep_const  = qog.iaep_const.replace('.b', np.NaN) ## no constitution

## Dropping psi_edate1 and psi_edate2 which are Exact date of the election and Exact date of the second election in a year.
qog = qog.drop(['psi_edate1', 'psi_edate2'], axis = 1)

column_subset = list(qog.columns) ## Getting a list of all columns

## Removing unit of analysis columns from the list of columns
for entry in ['ccode', 'cname', 'year', 'ccodealp', 'cname_year', 'ccodealp_year', 'ccodecow', 'ccodewb', 'version']:
    column_subset.remove(entry)

qog = qog.dropna(axis = 0, how = "all", subset = column_subset) ## Drop rows containing all NA's
qog = qog.dropna(axis = 1, how = "all") ## Drop columns containing all NA's
qog = qog.reset_index(drop = True) ## Resetting index
```

```python
qog2 = qog.drop(['ccode', 'cname', 'year', 'ccodealp', 'cname_year', 'ccodealp_year', 'version', 'ccodecow', 'ccodewb'], axis = 1) ## Dropping unnecessary columns for PCA
qog2 = qog2.astype(float) ## Changing dtype to float
qog2 = qog2.reset_index(drop = True) ## Resetting index

## Standardization of Variables
dont_normalize_cols_qog2 = [col for col in qog2 if np.isin(qog2[col].dropna().unique(), range(0,15)).all()] ## Getting list of variables that should not be standardized
dont_normalize_qog2 = qog2[dont_normalize_cols_qog2] ## storing columns that should not be standardize in another df to merge back later
normalize_qog2 = qog2.drop(dont_normalize_cols_qog2, axis = 1) ## standardize

scaler = StandardScaler() ## To standardize all variables
qog3 = pd.DataFrame(scaler.fit_transform(normalize_qog2), ## Standardizing
                    columns = normalize_qog2.columns)
qog3 = pd.concat([qog3,dont_normalize_qog2],axis=1,sort=False) ## Concatenating unstandardized variables
qog3 = qog3.reset_index(drop = True) ## Resetting index
```

Checking for missing values in variables.

```python
fig = plt.figure()
plt.scatter(list(dict(qog3.isna().sum()).keys()), list(dict(qog3.isna().sum()).values()), color = 'black', alpha = 0.2)
plt.ylim(0,17000)
plt.axhline(15324, color ='darkred')
titlefont = {'fontname':'Times New Roman', 'size': 15}
subtitlefont = {'fontname':'Times New Roman', 'size': 10}
plt.xlabel("Different Variables", **subtitlefont)
plt.ylabel("Number of Missing Entries", **subtitlefont)
plt.title("Total Missing Entries", **titlefont)
plt.annotate('Total Observations', xy=(0, 15324), xytext=(-0.05, 3300),
            arrowprops=dict(facecolor='black', shrink = 0.05, width = 1.2))
plt.tick_params(axis = "x", which = "both", bottom = False, top = False, labelbottom=False)
fig.savefig("Figures/figure3.png", figsize = (10,5), dpi = 600, edgecolor = 'b')
```

Creating target array for AHC

```python
x = scored_dataframe.drop(['cname', 'year', 'ccodealp', 'cname_year', 'ccodealp_year'], axis = 1) # Creating target array for AHC
x.head()
```

```python
fig = plt.figure(figsize = (7,5))
dendrogram = dendrogram(linkage(x, method='ward'))
titlefont = {'fontname':'Times New Roman', 'size': 15}
subtitlefont = {'fontname':'Times New Roman', 'size': 10}
plt.xlabel("Different Variables", **subtitlefont)
plt.title("Dendrogram for Agglomerative Hierarchical Cluster Analysis using Ward linkages", **titlefont)
plt.tick_params(axis = "both", which = "both", bottom = False, top = False, labelbottom=False)
fig.savefig("Figures/figure7.png", figsize = (10,5), edgecolor = 'b')
```

```python
# Agglomerative Hierarchical clustering

ahc = AgglomerativeClustering(linkage='ward', n_clusters = 2)
ahc_clusters2 = ahc.fit_predict(x)

ahc = AgglomerativeClustering(linkage='ward', n_clusters = 4)
ahc_clusters4 = ahc.fit_predict(x)

ahc = AgglomerativeClustering(linkage='ward', n_clusters = 6)
ahc_clusters6 = ahc.fit_predict(x)
```

```python
scored_dataframe['cluster_2'] = ahc_clusters2
scored_dataframe['cluster_4'] = ahc_clusters4
scored_dataframe['cluster_6'] = ahc_clusters6
```
